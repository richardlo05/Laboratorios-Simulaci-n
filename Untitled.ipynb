```python
def congruencial(a,b,m,n,Xo):
    vector=[]
    for i in range(n):
        Xn=(a*Xo+b)%m
        Xo=Xn
        Un=Xn/float(m)
        vector.append(Un)
    return vector
v = congruencial(3,7,29,20,3) 
print (v)
```
